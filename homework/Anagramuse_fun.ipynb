```python
def Anagram(a,b):
    for ch in a:
        if ch in b:
            b=b.replace(ch,"",1)
        else:
            print("Not Anagram")
            return
    if b == "":
        print("Anagram")
    else:
        print("Not Anagram")
x=input()
y=input()
Anagram(x,y)

            
    
    
```

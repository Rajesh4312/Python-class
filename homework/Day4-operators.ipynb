```python
# Arithmetic Operators
a=8
b=9
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a//b)
print(a**b)
```

```python
# Assignment Operators
a=5
a+=3
a-=3
a*=4
a/=2
```

```python
# Comparison Operators
a=5
b=3
print(a==b)
print(a>b)
print(a<b)
print(a!=b)
print(a>=b)
print(a<=b)
```

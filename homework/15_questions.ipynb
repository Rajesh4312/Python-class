```python
# Anagram

word1 = input().lower()
word2 = input().lower()

if len(word1) != len(word2):
    print("Not anagram")

else:
    count = {}

    for char in word1:
        count[char] = count.get(char, 0) + 1

    for char in word2:
        if char in count:
            count[char] -= 1
            if count[char] < 0:   # 🔥 important fix
                print("Not anagram")
                break
        else:
            print("Not anagram")
            break
    else:
        print("Anagram")
    
    
```

```python
s="madam"
rev=""
for i in s:
    rev = i+rev
if s==rev:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
n=int(input())
if n <=1:
    print("Not prime")
else:
    for i in range(2,n):
        if n % i == 0:
            print("Notprime")
            break
    else:
         print("prime")
```

```python
# Fibonacci Series
num=int(input())
a=0
b=1
for i in range(num):
    print(a,end="")
    c=a+b
    a=b
    b=c
```

```python
l=[2,4,67,94,1,5]
max=l[0]
min=l[0]
for i in l:
    if i > max:
        max=i
    if i < min:
        min=i
print(max)
print(min)
        
```

```python
# smallest and largest
l=[2,4,67,94,1,5] 
k=len(l)
for i in range(k):
    for j in range(0,k-i-1):
        if l[j] > l[j+1]:
            temp=l[j]
            l[j]=l[j+1]
            l[j+1]=temp
print(l)
print(l[1])
print(l[-2])

            
```

```python
k=101
original=k
reverse=0
while k > 0:
    digit= k % 10
    reverse = reverse * 10 +digit
    k=k//10

if original == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
    
    
```

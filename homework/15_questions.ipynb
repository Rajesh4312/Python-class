```python
# Anagram

word1 = input().lower()
word2 = input().lower()

if len(word1) != len(word2):
    print("Not anagram")

else:
    count = {}

    for char in word1:
        count[char] = count.get(char, 0) + 1

    for char in word2:
        if char in count:
            count[char] -= 1
            if count[char] < 0:   # 🔥 important fix
                print("Not anagram")
                break
        else:
            print("Not anagram")
            break
    else:
        print("Anagram")
    
    
```

```python
l = [2,4,67,94,1,5]

n = len(l)

# Ascending order
for i in range(n):
    for j in range(0,n-i-1):
        if l[j] > l[j+1]:
            temp = l[j]
            l[j] = l[j+1]
            l[j+1] = temp

print("Ascending Order:", l)

# Descending order
print("Descending Order:", end=" ")
for i in range(n-1, -1, -1):
    print(l[i], end=" ")
```

```python
# anagram

word1 = input().lower()
word2 = input().lower()

if len(word1) != len(word2):
    print("Not anagram")

else:
    count = {}

    for char in word1:
        count[char] = count.get(char, 0) + 1

    for char in word2:
        if char in count:
            count[char] -= 1
            if count[char] < 0:   # 🔥 important fix
                print("Not anagram")
                break
        else:
            print("Not anagram")
            break
    else:
        print("Anagram")
```

```python
# Find 2nd Largest and 2nd Smallest Number
n = [2, 36, 87, 98, 2]
k = len(n)

for i in range(k):
    for j in range(0, k - i - 1):
        if n[j] > n[j + 1]:
            temp = n[j]
            n[j] = n[j + 1]
            n[j + 1] = temp  
print(n)
print("2nd Smallest:", n[1])
print("2nd Largest:", n[k - 2])
    
```

```python
num = 153
order = len(str(num))
total = 0
temp = num

while temp > 0:
    digit = temp % 10
    total += digit ** order
    temp //= 10

if total == num:
    print("Armstrong")
else:
    print("Not Armstrong")
```

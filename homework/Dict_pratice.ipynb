```python
employees = {
    "Raj": (101, 30000),
    "Arun": (102, 35000),
    "Kumar": (103, 40000)
}

while True:

    print("\n1. Add Employee")
    print("2. Edit Employee Name")
    print("3. Delete Employee")
    print("4. Display Employees")
    print("5. Exit")

    choice = int(input("Enter a choice: "))

    if choice == 1:
        name = input("Enter name: ")
        emp_id = int(input("Enter ID: "))
        salary = int(input("Enter salary: "))
        employees[name] = (emp_id, salary)
        print("Employee added")

    elif choice == 2:
        old_name = input("Enter old name: ")

        if old_name in employees:
            new_name = input("Enter new name: ")
            employees[new_name] = employees.pop(old_name)
            print("Name updated")
        else:
            print("Record not found")

    elif choice == 3:
        old_name = input("Enter name: ")

        if old_name in employees:
            del employees[old_name]
            print("Deleted")
        else:
            print("Not found")

    elif choice == 4:
        print("\nEmployee Records")
        for name, data in employees.items():
            print("Name:", name, "ID:", data[0], "Salary:", data[1])

    elif choice == 5:
        break

    else:
        print("Invalid Choice")
```

```python
empdt={
    "1": {"name":"Dinesh",   "id": 1,  "salary":100000},
    "2": {"name": "Rahul",   "id": 2,  "salary": 85000},
    "3": {"name": "Priya",   "id": 3,  "salary": 92000},
    "4": {"name": "Karthik", "id": 4,  "salary": 78000},
    "5": {"name": "Meena",   "id": 5,  "salary": 88000},
    "6": {"name": "Arun",    "id": 6,  "salary": 95000},
    "7": {"name": "Anu",     "id": 7,  "salary": 73000},
    "8": {"name": "Vijay",   "id": 8,  "salary": 99000},
    "9": {"name": "Sneha",   "id": 9,  "salary": 87000},
    "10": {"name": "Ravi",   "id": 10, "salary": 91000} 
}

task1 =input(
    """Select a Key To Perform Task
    Key a for Add a New Record
    Key u for Update an Existing Record
    Key d for Delete an Record
    """)
# ADDING
if task1 =="a":
    ipn=input("Enter The Name Of The Employee:  ")
    ipi=int(input("Enter The Id Of The Employee:  "))
    ips=int(input("Enter The Salary Of The Employee:  "))
    new_key = str(len(empdt)+1)
    empdt[new_key]={"name":ipn,"id": ipi,"salary":ips}
    print("Record Successfully Added")


    
# UPDATION
elif task1=="u":
    ipnu=input("Enter The Name Of The Employee:  ")
    for key , value in empdt.items():
        if value["name"].lower()==ipnu.lower():
            up=input("""Select The correct keys To Perform The Updateion 
            For Name Press n 
            For Salary Press s
            For Both Press b  :""")
            if up =="n":
                new_name=input("Enter The New Name :")
                value["name"] = new_name
                print("Record Updated SuccessFully")
            elif up =="s":
                new_salary=int(input("Enter The New Salary :"))
                value["salary"] = new_salary
                print("Record Updated SuccessFully")
            elif up =="b":
                new_name=input("Enter The New Name :")
                new_salary=int(input("Enter The New Salary :"))
                value["name"] = new_name
                value["salary"] = new_salary
                print("Record Updated SuccessFully")
                break
            else:
                print("Record Not Found")
        

            
        
# DELETE
elif task1 =="d":
    ds = input("Enter The Name :")
    for key , value in empdt.items():
        if value["name"].lower()==ds.lower():
            del empdt[key]
            print("Record Deleted Successfully")
            break
else:
    print("Record Not Found")

print("\nFinal Data:\n")
for key, value in empdt.items():
    print(f"Key: {key} | Name: {value['name']} | ID: {value['id']} | Salary: {value['salary']}")
    
```
